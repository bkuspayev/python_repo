```python
import pandas as pd
from scipy.ndimage import gaussian_filter
import matplotsoccer
import matplotlib.pyplot as plt
```

```python
game = pd.read_excel('soccer.xlsx', 1)
```

```python
teams = game.team.unique()
```

```python
for i in range(len(teams)):
    print(i, teams[i])
```

```python
team = teams[int(input())]
matplotsoccer.field("green",figsize=8, show=False)
x = game[game.team == f'{team}'].pos_x
y = game[game.team == f'{team}'].pos_y
plt.scatter(x, y)
plt.axis("on")
plt.show()
print(team)
```

```python
hm = matplotsoccer.count(x,y,n=25,m=25) # Construct a 25x25 heatmap from x,y-coordinates
hm = gaussian_filter(hm,1) # blur the heatmap
matplotsoccer.heatmap(hm) # plot the heatmap
```
