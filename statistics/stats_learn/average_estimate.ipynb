```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
mu = 100
sigma = 10
```

```python
population = np.random.normal(mu, sigma, 10000).round(2)
```

```python
sns.distplot(population, kde=False)
```

```python
population.std()
```

```python
sample_size = 30
```

```python
sample = np.random.choice(population, sample_size, False)
```

```python
sample.mean()
```

```python
n = 10000
```

```python
sample_means = []
```

```python
for i in range(n):
    sample = np.random.choice(population, sample_size, False)
    sample_means.append(sample.mean())
```

```python
sample_means
```

```python
sns.distplot(sample_means, kde=False)
```

```python
np.mean(sample_means)
```

```python
se = population.std() / (sample_size ** 0.5)
se
```

```python
np.std(sample_means)
```

```python
population.mean() - 1.96 * se # lower bound
```

```python
population.mean() + 1.96 * se # upper bound
```
