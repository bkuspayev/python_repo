```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
mu = 100
sigma = 10
population = np.random.normal(mu, sigma, 10000).round(2)
sns.distplot(population, kde=False)
```

```python
population.std()
```

```python
sample_size = 10
sample = np.random.choice(population, sample_size, False)
sample.mean()
```

```python
se = population.std() / (sample_size ** 0.5)
se
```

```python
n = 10000
sample_means = []
for i in range(n):
    sample = np.random.choice(population, sample_size, False)
    sample_means.append(sample.mean())
```

```python
sns.distplot(sample_means, kde=False)
```

```python
population.mean()
```

```python
z_scores = []
```

```python
n = 10000
for i in range(n):
    sample = np.random.choice(population, sample_size, False)
    sample_means.append(sample.mean())

    z_score = (sample.mean() - population.mean()) / se
    z_scores.append(z_score)
```

```python
z_scores
```

```python
sns.distplot(z_scores, kde=False)
```

```python
t_scores = []
n = 10000
for i in range(n):
    sample = np.random.choice(population, sample_size, False)
    sample_means.append(sample.mean())
    sample_std = sample.std()
    se_ = sample_std / sample_size ** 0.5
    t_score = (sample.mean() - population.mean()) / se_
    t_scores.append(t_score)
```

```python
sns.distplot(t_scores, kde=False)
```
